```python
from pathlib import Path
from sssom.parsers import parse_sssom_table
from sssom.util import compare_dataframes
import pandas as pd
```

```python
proj_dir = Path("__file__").resolve().parents[1]
data_dir = proj_dir / "data"
input_sssom = data_dir / "mondo_omim_ordo_do.sssom.tsv"
comparison_dir = proj_dir / "omim-ordo-do" / "boomer_output" / "mondo_omim_ordo_do"/"comparison"
output_sssom = comparison_dir / "boomer.sssom.tsv"
```

```python
%%time
input_msdf = parse_sssom_table(input_sssom)
input_msdf.df.head()
```

```python
%%time
output_msdf = parse_sssom_table(output_sssom)
output_msdf.df.head()
```

```python
%%time
combo = compare_dataframes(input_msdf.df, output_msdf.df)
```

```python
combo.combined_dataframe.drop_duplicates()
```

```python
combo_unique = combo.combined_dataframe.drop_duplicates()

uncommon_df = combo_unique[~(combo_unique["comment"] == "COMMON_TO_BOTH")]

uncommon_df
```

```python
uncommon_df.to_csv(comparison_dir/"uncommon.sssom.tsv", sep="\t", index=False)
```
